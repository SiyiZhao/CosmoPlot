# linear matter power spectrum

Here we show how linear matter power spectrum changes with the vanilla cosmological parameters.

```python
import numpy as np
from colossus.cosmology import cosmology
from matplotlib import pyplot as plt
```

```python
z = 0
k = np.logspace(-4, 2, 1000)
```

```python
cosmo = cosmology.setCosmology('planck18')
pk = cosmo.matterPowerSpectrum(k, z)
```

```python
cosmo.Om0, cosmo.sigma8
```

## Omega_m

```python
cosmo_om = cosmology.setCosmology('planck18', {'Om0': 0.1})
pk_om = cosmo_om.matterPowerSpectrum(k, z)
cosmo_om_p = cosmology.setCosmology('planck18', {'Om0': 0.5})
pk_om_p = cosmo_om_p.matterPowerSpectrum(k, z)
```

```python
plt.figure(dpi=150)
plt.loglog(k, pk, label=r'Planck18 $\Omega_m = 0.3111$')
plt.loglog(k, pk_om, label=r'$\Omega_m = 0.1$')
plt.loglog(k, pk_om_p, label=r'$\Omega_m = 0.5$')
plt.xlabel(r'$k\,[h/\mathrm{Mpc}]$')
plt.ylabel(r'$P(k)\,[(\mathrm{Mpc}/h)^3]$')
plt.legend()
plt.show()
```

## sigma8

```python
cosmo_s8 = cosmology.setCosmology('planck18', {'sigma8': 0.5})
pk_s8 = cosmo_s8.matterPowerSpectrum(k, z)
cosmo_s8_p = cosmology.setCosmology('planck18', {'sigma8': 1.0})
pk_s8_p = cosmo_s8_p.matterPowerSpectrum(k, z)
```

```python
plt.figure(dpi=150)
plt.loglog(k, pk, label=r'Planck18 $\sigma_8 = 0.8102$')
plt.loglog(k, pk_s8, label=r'$\sigma_8 = 0.5$')
plt.loglog(k, pk_s8_p, label=r'$\sigma_8 = 1.0$')
plt.xlabel(r'$k\,[h/\mathrm{Mpc}]$')
plt.ylabel(r'$P(k)\,[(\mathrm{Mpc}/h)^3]$')
plt.legend()
plt.show()
```

## H0

```python
cosmo.H0
```

```python
cosmo_h0 = cosmology.setCosmology('planck18', {'H0': 50})
pk_h0 = cosmo_h0.matterPowerSpectrum(k, z)
cosmo_h0_p = cosmology.setCosmology('planck18', {'H0': 100})
pk_h0_p = cosmo_h0_p.matterPowerSpectrum(k, z)
```

```python
plt.figure(dpi=150)
plt.loglog(k, pk, label=r'Planck18 $H_0 = 67.66$')
plt.loglog(k, pk_h0, label=r'$H_0 = 50$')
plt.loglog(k, pk_h0_p, label=r'$H_0 = 100$')
plt.xlabel(r'$k\,[h/\mathrm{Mpc}]$')
plt.ylabel(r'$P(k)\,[(\mathrm{Mpc}/h)^3]$')
plt.legend()
plt.show()
```
